# ..."Series"

```python
import numpy as np
data=np.array([101,102,103,104])
s=pd.Series(data)
print(s,"\n")
print(s[1])
s[1]=100    #TO modify
s[4]=200     # to add new value
print(s)
```

```python
s=pd.Series(data,index=['a','b','c','d'])
print(s)
print(s['b'])
print(s[1])
```

```python
list1=np.arange(1001,1005,1)
s = pd.Series(data,index=list1)
print(s)
print(s[1002])
#print(s[1])  #over write the default index number "key error" will occur
```

```python
data = {'a':0.,'b':1.,'c':2.,'d':3.}
s = pd.Series(data)
print(s)
print(s['a'],"\n")
print(s[-4:])
```

```python
#Generic data set="Monetry loss(if any mistakes)"
#Sensitive data set="Human loss(if any mistakes)"
#if we have null values>50% then we drop the variable(for generic dataset).
# missing valuse>75% null value then we drop the variable(sensitive dataset).  

#Generic dataset
half_count = len(titanic_df)/2
titanic_df = titanic_df.dropna(thresh=half_count,axis=1)
print(titanic_df.isnull().sum())


"""
half_count=0.75*len(titanic_df)
print(half_count)
"""#Sensitive dataset
```

```python
titanic_df=titanic_df.drop(['Ticket','Fare'],axis=1)
titanic_df.head()
```

```python
#1. discussing with client
#2. default value
#3. Measure of Center Tendency
#4. Predicting the values (classsifiction technique)
#      Age!=na (non misssing value=train)
#      Age=na (missing value=test)
```

```python
#handling mising values in Embarked column
titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0],inplace=True)
titanic_df['Embarked']=titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0])
```
